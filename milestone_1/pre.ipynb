### Why Rescale Data?

### Gradient Descent-based models like neural nets: Rescaling ensures all features contribute equally, avoiding bias toward large-magnitude features.


### Regularization-based models like ridge regression: Features with larger values dominate, leading to incorrect penalty application.


### Distance-based models like SVM: Similair to previous examples, it prevents dominance by large-scale features, ensuring meaningful distance calculations.

# Baseline Models 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

X = df_scaled.drop(columns=['target'])  
y = df_scaled['target']  

# Initialize models
models = {
    'Ridge Regression': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Define k-fold cross validation with fixed random seed for consistency
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Dictionary to store results
results = {}
model_in_errors = {}
model_out_errors = {}
model_in_point_errors = {}
model_out_point_errors = {}


print("Performing k-fold cross-validation with consistent splits across all models...")

# Iterate through each fold with the same splits for all models
fold_idx = 0
for train_idx, test_idx in kf.split(X):
    fold_idx += 1
    print(f"Processing fold {fold_idx}/{k_folds}...")
    
    # Split data for this fold
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # Apply each model to the same fold
    for name, model in models.items():
        # Initialize storage for this model if not already done
        if name not in model_in_errors:
            model_in_errors[name] = []
            model_out_errors[name] = []
            model_in_point_errors[name] = []
            model_out_point_errors[name] = []
        
        # Fit model
        model.fit(X_train, y_train)
        
        # Predict
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)
        
        # Calculate fold errors (MSE)
        in_err = mean_squared_error(y_train, y_train_pred)
        out_err = mean_squared_error(y_test, y_test_pred)
        
        # Store fold errors for t-tests
        model_in_errors[name].append(in_err)
        model_out_errors[name].append(out_err)
        
        # Calculate and store point-wise squared errors for KDE
        train_point_errors = (y_train - y_train_pred) ** 2
        test_point_errors = (y_test - y_test_pred) ** 2
        
        model_in_point_errors[name].extend(train_point_errors)
        model_out_point_errors[name].extend(test_point_errors)

# Calculate mean errors for each model
for name in models.keys():
    results[name] = {
        'in_sample_mean': np.mean(model_in_errors[name]),
        'out_sample_mean': np.mean(model_out_errors[name]),
        'in_sample_std': np.std(model_in_errors[name]),
        'out_sample_std': np.std(model_out_errors[name])
    }
    
    print(f"\n{name}:")
    print(f"  In-sample MSE: {results[name]['in_sample_mean']:.6f} ± {results[name]['in_sample_std']:.6f}")
    print(f"  Out-of-sample MSE: {results[name]['out_sample_mean']:.6f} ± {results[name]['out_sample_std']:.6f}")

# Paired t-tests for in-sample errors
print("\n=== Paired t-tests for in-sample errors ===")
model_names = list(models.keys())
for i in range(len(model_names)):
    for j in range(i+1, len(model_names)):
        model1 = model_names[i]
        model2 = model_names[j]
        t_stat, p_val = stats.ttest_rel(
            model_in_errors[model1],
            model_in_errors[model2]
        )
        print(f"{model1} vs {model2}: t-statistic = {t_stat:.8f}, p-value = {p_val:.8f}")

# Paired t-tests for out-of-sample errors
print("\n=== Paired t-tests for out-of-sample errors ===")
for i in range(len(model_names)):
    for j in range(i+1, len(model_names)):
        model1 = model_names[i]
        model2 = model_names[j]
        t_stat, p_val = stats.ttest_rel(
            model_out_errors[model1],
            model_out_errors[model2]
        )
        print(f"{model1} vs {model2}: t-statistic = {t_stat:.7f}, p-value = {p_val:.7f}")

# Create KDE plots for error distributions
plt.figure(figsize=(15, 10))

# Function to create KDE plot
def plot_kde(ax, error_dict, title):
    for name, errors in error_dict.items():
        # Convert to numpy array and reshape for KDE
        errors_array = np.array(errors).reshape(-1, 1)
        
        # Determine appropriate bandwidth - Scott's rule is a good default
        bandwidth = 1.06 * np.std(errors_array) * len(errors_array) ** (-0.2)
        
        # Handle very small or zero bandwidth
        if bandwidth < 1e-6:
            bandwidth = 0.01
            
        # Fit KDE
        x = np.linspace(0, np.percentile(errors_array, 99), 1000).reshape(-1, 1)
        
        from sklearn.neighbors import KernelDensity
        kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth).fit(errors_array)
        log_dens = kde.score_samples(x)
        ax.plot(x, np.exp(log_dens), label=name)

    ax.set_title(title)
    ax.set_xlabel('Squared Error')
    ax.set_ylabel('Density')
    ax.legend()
    ax.grid(True, alpha=0.3)

# KDE for in-sample errors
plt.subplot(2, 1, 1)
plot_kde(plt.gca(), model_in_point_errors, 'In-Sample Error Distribution (KDE)')

# KDE for out-of-sample errors
plt.subplot(2, 1, 2)
plot_kde(plt.gca(), model_out_point_errors, 'Out-of-Sample Error Distribution (KDE)')

plt.tight_layout()
plt.savefig('error_distributions.png')
plt.show()

# Create a bar chart comparing in-sample and out-of-sample errors
plt.figure(figsize=(12, 6))
x = np.arange(len(model_names))
width = 0.35

in_errors = [results[name]['in_sample_mean'] for name in model_names]
out_errors = [results[name]['out_sample_mean'] for name in model_names]
in_std = [results[name]['in_sample_std'] for name in model_names]
out_std = [results[name]['out_sample_std'] for name in model_names]

plt.bar(x - width/2, in_errors, width, label='In-sample MSE', yerr=in_std, capsize=5)
plt.bar(x + width/2, out_errors, width, label='Out-of-sample MSE', yerr=out_std, capsize=5)

plt.xlabel('Models')
plt.ylabel('Mean Squared Error')
plt.title('In-sample vs Out-of-sample Mean Squared Error by Model')
plt.xticks(x, model_names)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('error_comparison.png')
plt.show()

# Rank models by out-of-sample performance
model_ranks = {name: rank for rank, name in 
               enumerate(sorted(results.keys(), key=lambda x: results[x]['out_sample_mean']))}

print("\n=== MODEL RANKING (based on out-of-sample MSE) ===")
for name, rank in sorted(model_ranks.items(), key=lambda x: x[1]):
    print(f"{rank+1}. {name}: {results[name]['out_sample_mean']:.6f}")

# Print comprehensive summary
print("\n=== COMPREHENSIVE SUMMARY ===")
print("\nMSE Comparison:")
for name in model_names:
    print(f"{name}:")
    print(f"  In-sample MSE: {results[name]['in_sample_mean']:.6f} ± {results[name]['in_sample_std']:.6f}")
    print(f"  Out-of-sample MSE: {results[name]['out_sample_mean']:.6f} ± {results[name]['out_sample_std']:.6f}")
   
    
```

```python
# One-tailed paired t-tests for out-of-sample errors
print("\n=== One-tailed paired t-tests (testing if first model is better) ===")

# Test if Ridge is better than Random Forest
t_stat, p_val = stats.ttest_rel(
    model_out_errors['Ridge Regression'],
    model_out_errors['Random Forest'],
    alternative='less'  # Test if Ridge has lower error
)
print(f"H₀: Ridge Regression is NOT better than Random Forest")
print(f"H₁: Ridge Regression is better than Random Forest")
print(f"t-statistic = {t_stat:.7f}, p-value = {p_val:.7f}")
print(f"Conclusion: {'Reject H₀' if p_val < 0.05 else 'Fail to reject H₀'}")
print()

# Test if Random Forest is better than Gradient Boosting
t_stat, p_val = stats.ttest_rel(
    model_out_errors['Random Forest'],
    model_out_errors['Gradient Boosting'],
    alternative='less'  # Test if Random Forest has lower error
)
print(f"H₀: Random Forest is NOT better than Gradient Boosting")
print(f"H₁: Random Forest is better than Gradient Boosting")
print(f"t-statistic = {t_stat:.7f}, p-value = {p_val:.7f}")
print(f"Conclusion: {'Reject H₀' if p_val < 0.05 else 'Fail to reject H₀'}")
print()

# Test if Random Forest is better than Gradient Boosting
t_stat, p_val = stats.ttest_rel(
    model_out_errors['Gradient Boosting'],
    model_out_errors['Random Forest'],
    alternative='less'  # Test if Gradient Boosting has lower error
)
print(f"H₀: Gradient Boosting is NOT better than Random Forest")
print(f"H₁: Gradient Boosting is better than Random Forest")
print(f"t-statistic = {t_stat:.7f}, p-value = {p_val:.7f}")
print(f"Conclusion: {'Reject H₀' if p_val < 0.05 else 'Fail to reject H₀'}")
print()

# Test if Gradient Boosting is better than Ridge
t_stat, p_val = stats.ttest_rel(
    model_out_errors['Gradient Boosting'],
    model_out_errors['Ridge Regression'],
    alternative='less'  # Test if Gradient Boosting has lower error
)
print(f"H₀: Gradient Boosting is NOT better than Ridge Regression")
print(f"H₁: Gradient Boosting is better than Ridge Regression")## Gradient Boosting is better than Ridge (error for Gradient Boosting is less than Ridge
print(f"t-statistic = {t_stat:.7f}, p-value = {p_val:.7f}")
print(f"Conclusion: {'Reject H₀' if p_val < 0.05 else 'Fail to reject H₀'}")
```
