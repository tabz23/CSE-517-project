```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Define models
models = {
    "RidgeCV": RidgeCV(alphas=np.logspace(-6, 6, 13)),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "GaussianProcessRegressor": GaussianProcessRegressor(kernel=RBF())
}

# Define k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store results
in_sample_errors = {}
out_of_sample_errors = {}

# Loss function: Mean Squared Error (MSE)

# Fit and evaluate models
for model_name, model in models.items():
    in_sample_errors_list = []
    out_of_sample_errors_list = []
    for train_index, test_index in kf.split(X_normalized):
        X_train, X_test = X_normalized[train_index], X_normalized[test_index]
        y_train, y_test = y_normalized[train_index], y_normalized[test_index]

        model.fit(X_train, y_train)

        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        in_sample_errors_list.append(mean_squared_error(y_train, y_train_pred))
        out_of_sample_errors_list.append(mean_squared_error(y_test, y_test_pred))

    in_sample_errors[model_name] = in_sample_errors_list
    out_of_sample_errors[model_name] = out_of_sample_errors_list
```

```python

import matplotlib.pyplot as plt
# Perform paired t-tests
model_names = list(models.keys())
for i in range(len(model_names)):
    for j in range(i + 1, len(model_names)):
        model1_name = model_names[i]
        model2_name = model_names[j]

        # In-sample errors
        t_statistic_in, p_value_in = stats.ttest_rel(
            in_sample_errors[model1_name], in_sample_errors[model2_name]
        )
        print(f"In-sample errors: Paired t-test between {model1_name} and {model2_name}")
        print(f"  t-statistic: {t_statistic_in:.3f}")
        print(f"  p-value: {p_value_in:.3f}")
        if p_value_in < 0.05:
            print("  Significant difference in in-sample errors.")
        else:
            print("  No significant difference in in-sample errors.")


        # Out-of-sample errors
        t_statistic_out, p_value_out = stats.ttest_rel(
            out_of_sample_errors[model1_name], out_of_sample_errors[model2_name]
        )
        print(f"Out-of-sample errors: Paired t-test between {model1_name} and {model2_name}")
        print(f"  t-statistic: {t_statistic_out:.3f}")
        print(f"  p-value: {p_value_out:.3f}")
        if p_value_out < 0.05:
            print("  Significant difference in out-of-sample errors.")
        else:
            print("  No significant difference in out-of-sample errors.")


# Kernel density estimates
for model_name in model_names:
    # In-sample errors
    plt.figure()
    sns.kdeplot(in_sample_errors[model_name], label=model_name)
    plt.xlabel("In-sample error")
    plt.ylabel("Density")
    plt.title(f"In-sample error distribution for {model_name}")
    plt.legend()
    plt.show()

    # Out-of-sample errors
    plt.figure()
    sns.kdeplot(out_of_sample_errors[model_name], label=model_name)
    plt.xlabel("Out-of-sample error")
    plt.ylabel("Density")
    plt.title(f"Out-of-sample error distribution for {model_name}")
    plt.legend()
    plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression



# K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X_normalized)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Calculate distances from cluster centers
distances = []
for i in range(len(X_normalized)):
    point = X_normalized[i]
    distance_to_center0 = np.linalg.norm(point - cluster_centers[0])
    distance_to_center1 = np.linalg.norm(point - cluster_centers[1])
    distances.append([distance_to_center0, distance_to_center1])

distances = np.array(distances)


# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(distances[:, 0], distances[:, 1], c=y_normalized, cmap='viridis') # Use color to represent target
plt.xlabel("Distance from Cluster Center 0")
plt.ylabel("Distance from Cluster Center 1")
plt.title("K-means Clustering Result")
plt.colorbar(label='Target Value (Normalized)')
plt.show()


# Linear Regression
X_distances = distances
y_distances = y_normalized
model = LinearRegression()
model.fit(X_distances, y_distances)
y_pred = model.predict(X_distances)


print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
